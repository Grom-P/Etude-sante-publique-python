<h3>2.Un fichier de disponibilité alimentaire pour l’année 2013 qui contient 17 variables :</h3><font></font>

* Le pays.
* Le produit.
* 4 variables de disponibilité alimentaire. Ces 4 variables sont exprimées dans des unités
bien spécifiques, qui sont précisées directement dans le nom de la variable. À noter
cependant que la notation kcal est un abus de langage, elle correspond à une calorie
réelle.

* Enfin, une variable type qui précise si le produit est d’origine animale ou végétale.

```python
# Trouver le nombre des doublons
dispo_al.duplicated(['Pays','Produit']).sum()
```

```python
# Pourcentage de valeurs manquantes
dispo_al.isnull().mean()
```

```python
#affiche les infos
dispo_al.info()
```

```python
#affiche le nombre de valeurs manquante
print(dispo_al.isnull().sum())
```

```python
#remplie les valeurs manquante par 0
dispo_al.fillna(0, inplace=True)

#affiche le nombre de valeurs manquante
print(dispo_al.isnull().sum())
```

```python
#changement du type des colonnes en entier
col_types = {'Aliments pour animaux':'int64',
              'Disponibilité intérieur':'int64', 'Exportations':'int64',
       'Importations':'int64', 'Nourriture humaine':'int64', 'Pertes':'int64',
       'Production':'int64', 'Semences':'int64', 'Traitement':'int64',
       'Variation de stock':'int64'}
dispo_al = dispo_al.astype(col_types)
```

<h3>4.Un fichier d’insécurité alimentaire, qui inclut :</h3><font></font>

* Le pays.
* Intervalle d’années.
* Le nombre de personnes en sous-alimentation en millions d’habitants (ex. : pour
l’intervalle 2012-2014, le nombre de personnes correspond à la moyenne sur ces 3
années, on peut le résumer en 2013).

```python
#creation du dataframe sous_nut
sous_nut = pd.read_csv('sous_nutrition.csv') 

#affichage des 5 premère lignes
sous_nut.head(1)
```

```python
#renommage des colonnes
sous_nut= sous_nut.rename(columns={"Zone": "Pays","Valeur":"Sous nutrition"})
```

```python
#mappage des valeurs de la colonne
sous_nut['Année']=sous_nut['Année'].map({'2012-2014':'2013','2013-2015':'2014','2014-2016':'2015',
                                         '2015-2017':'2016','2016-2018':'2017','2017-2019':'2018'})
sous_nut.head()
```

```python
#affiche le nombre de valeur manquante
print(sous_nut.isnull().sum())
```

```python
# Trouver le nombre des doublons
sous_nut.duplicated(['Pays','Année']).sum()
```

```python
#affiches les infos
sous_nut.info()
```

```python
#affiches les valeurs aberrante
sous_nut.describe(include='all')
```

```python
#converti en numérique et si erreur transforme en valeur manquante les valeurs <0.1
sous_nut['Sous nutrition'] = pd.to_numeric(sous_nut['Sous nutrition'], errors='coerce')

sous_nut.head()
```

```python
#remplissage des valeurs manquante par un 0
sous_nut=sous_nut.fillna(0)

#affichage du nombre de valeur manquante
print(sous_nut.isnull().sum())
```

```python
#convertir en floattant
sous_nut['Sous nutrition']=sous_nut['Sous nutrition'].astype("float")

#multiplie par 1milliion
sous_nut['Sous nutrition']=sous_nut['Sous nutrition']*1000000
```

```python
#convertir les colonnes en entier
sous_nut['Année']=sous_nut['Année'].astype("int")

sous_nut['Sous nutrition']=sous_nut['Sous nutrition'].astype("int")

sous_nut.info()
```

```python
#création d'un dataframe des valeurs sous_nut de l'année 2017
sous_nutrition_2017 = sous_nut.loc[sous_nut['Année'] == 2017]
```

```python
sous_nutrition_2017.head(3) 
```
